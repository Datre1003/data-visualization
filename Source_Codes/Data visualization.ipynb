### 2. Trả lời câu hỏi dựa vào dataset:

#### 2.1 Mối quan hệ giữa tình trạng hôn nhân và quá trình học tập?

```python
selected = ['Marital status', 'Target']
df_target = df[selected]

# Create dataframe with 4 column:
new_df = pd.DataFrame()

# Create column according to categories in Target:
for target_type in df_target["Target"].unique():
    new_df[target_type] = (df_target["Target"] == target_type).astype(int)

# Add column "Marital status" into dataframe:
new_df["Marital status"] = df_target["Marital status"]

result = new_df.groupby("Marital status").sum().reset_index()

# Decode Marital status
result['Marital status'] = result['Marital status'].map({1: "Single", 2: "Married", 3: "Widower", 4: "Divorced", 5: "Facto union", 6: "Legally seperated"})


result.plot(x = 'Marital status', kind = 'bar', stacked = True, title = 'Martial status by Target')
plt.show()
```

##### Nhận xét:
- Tình trạng hôn nhân có ảnh hưởng khá nhiều đến quá trình học tập của học sinh, sinh viên.
- Ta có thể thấy rằng, nhóm độc thân chiếm phần lớn số lượng, lí do có thể là do họ muốn tập trung cho sự nghiệp trước khi muốn lập gia đình đồng thời họ cũng có nhiều thời gian dành cho học tập hơn, số lượng có thể tốt nghiệp trong nhóm người này cũng chiếm tỉ lệ khá cao.
- Đối với nhóm người đã kết hôn, số lượng ít hơn, bởi vì đa phần những người thuộc nhóm này đa phần đã có việc làm, thu nhập ổn định, và họ dành phần lớn thời gian để chăm sóc cho gia đình hơn là tiếp tục học tập, tỉ lệ tốt nghiệp và bỏ học của nhóm người này khá cân bằng.
- Đối với những nhóm khác, chỉ chiếm một phần rất nhỏ, nhưng ta cũng có thể thấy rằng, tình trạng mối quan hệ ít nhiều ảnh hưởng đến quá trình học tập. Nếu đó là một mối quan hệ lành mạnh, đem lại cho bạn tinh thần thoải mái, phấn chấn thì đó là một điều tốt, nhưng nếu đó là một mối quan hệ độc hại, nó sẽ ảnh hưởng tiêu cực đến sức khỏe tinh thần, dẫn đến kết quả học tập không như mong muốn.

#### 2.2 Những học sinh với một khoản nợ trong mình liệu có những áp lực, khó khăn gì ảnh hưởng đến quá trình học tập không?

```python
debt = df['Debtor']

# Tính toán tỉ lệ %:
grad_rate_debtor = df[df['Debtor'] == 1]['Target'].value_counts(normalize=True)['Graduate']
drop_rate_debtor = df[df['Debtor'] == 1]['Target'].value_counts(normalize=True)['Dropout']
enroll_rate_debtor = df[df['Debtor'] == 1]['Target'].value_counts(normalize=True)['Enrolled']
grad_rate_no_debtor = df[df['Debtor'] == 0]['Target'].value_counts(normalize=True)['Graduate']
drop_rate_no_debtor = df[df['Debtor'] == 0]['Target'].value_counts(normalize=True)['Dropout']
enroll_rate_no_debtor = df[df['Debtor'] == 0]['Target'].value_counts(normalize=True)['Enrolled']

# Dữ liệu để vẽ biểu đồ
labels = ['Graduate', 'Dropout', 'Enrolled']
sizes_debt = [grad_rate_debtor, drop_rate_debtor, enroll_rate_debtor]
sizes_no_debt = [grad_rate_no_debtor, drop_rate_no_debtor, enroll_rate_no_debtor]
colors = ['#8de5a1', '#ff9999', '#a1c9f4']
explode = (0.1, 0, 0)  

# Chia khung hình ra làm 2
fig, axes = plt.subplots(1, 2, figsize = (16, 6))

# Biểu đồ tròn 1
axes[0].pie(sizes_debt, explode=explode,  colors=colors, autopct='%1.1f%%', startangle=140)
axes[0].set_title('Sinh viên có nợ')
axes[0].axis('equal')

# Biểu đồ tròn 2
axes[1].pie(sizes_no_debt, explode=explode,  colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Sinh viên không nợ')
axes[1].axis('equal')

# Đặt tên chung cho cả hai biểu đồ
plt.suptitle('So sánh tỉ lệ nghỉ học giữa hai loại sinh viên')

# Tạo chú thích màu
plt.legend(labels, loc='lower left', bbox_to_anchor=(0.5, -0.1), ncol=3, frameon=False, prop={'size': 8}, handletextpad=0.5, columnspacing=1)
plt.show()
```

**Nhận xét:**
- Qua 2 biểu đồ trên, ta có thể nhận thức rõ ràng sự khác biệt rất lớn giữa 2 kiểu sinh viên.
- Những sinh viên không có nợ có tỉ lệ tốt nghiệp rất cao 53.8% gấp gần 2 lần tỉ lệ bỏ học giữa chừng là 28.3%.
- Ngược lại, những sinh viên có một khoản nợ lại có tỉ lệ bỏ học giữa chừng đáng báo động là 62%, gấp 3 lần tỉ lệ sinh viên có thể tốt nghiệp được 20.1%.
- Qua đó, ta có thể thấy được rằng, những viên có một khoản mợ, mang trong mình một áp lực khá lớn nên quá trình học tập có thể không tốt, dẫn đến bỏ học, cũng có những học sinh cố gắng vượt qua được áp lực để tốt nghiệp, đạt được thành tích tốt nhưng tỉ lệ này lại khá thấp.


#### 2.3 Tuổi tác có ảnh hưởng như thế nào đến quá trình học tập của học sinh, sinh viên?
